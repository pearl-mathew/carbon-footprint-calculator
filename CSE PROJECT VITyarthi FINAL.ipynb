```python
#CSE VITyarthi project - PEARL MATHEW
import tkinter as tk
from tkinter import messagebox

def calculate_carbon_footprint():
    try:
        electricity_emission_factor = 0.5  # kg CO2 per kWh
        fuel_emission_factor = 2.3         # kg CO2 per liter
        transport_emission_factor = 0.5    # kg CO2 per km

        # Getting input values with proper validation
        electricity_text = entry_electricity.get().strip()
        fuel_text = entry_fuel.get().strip()
        transport_text = entry_transport.get().strip()

        # Convert to float, use 0 if empty
        electricity_usage = float(electricity_text) if electricity_text else 0.0
        fuel_usage = float(fuel_text) if fuel_text else 0.0
        transport_usage = float(transport_text) if transport_text else 0.0

        # Calculating emission
        electricity_emission = electricity_usage * electricity_emission_factor
        fuel_emission = fuel_usage * fuel_emission_factor
        transport_emission = transport_usage * transport_emission_factor

        # Total carbon footprint sum
        total_emission = electricity_emission + fuel_emission + transport_emission

        # Update the result label
        label_result.config(text=f"Total Carbon Footprint: {total_emission:.2f} Kg CO2")
        print(f"DEBUG: Calculated total: {total_emission}")  # Debug line
        
    except ValueError as e:
        messagebox.showerror("Input Error", "Please enter valid numbers only!")  
        print(f"DEBUG: Error - {e}")  # Debug line
        reset_fields()

def reset_fields():

    entry_electricity.delete(0, tk.END)
    entry_fuel.delete(0, tk.END)
    entry_transport.delete(0, tk.END)
    label_result.config(text="")
    entry_electricity.focus()  

def confirm_exit():
    if messagebox.askyesno("Exit", "Are you sure you want to exit?"):
        root.quit()  
        root.destroy()

# Create the main window
root = tk.Tk()
root.title("Carbon Footprint Calculator")
root.geometry("1200x800")

frame = tk.Frame(root)
frame.pack(expand=True, fill='both', padx=20, pady=20)

tk.Label(frame, text="Carbon Footprint Calculator",
         font=("Arial", 36, "bold", "underline"), pady=10).pack()

tk.Label(frame, text="Electricity Usage (kWh):", font=("Arial", 24, "bold")).pack(pady=5)
entry_electricity = tk.Entry(frame, font=("Arial", 24), justify="center", width=15)
entry_electricity.pack(pady=5)

tk.Label(frame, text="Fuel Usage (liters):", font=("Arial", 24, "bold")).pack(pady=5)
entry_fuel = tk.Entry(frame, font=("Arial", 24), justify="center", width=15)
entry_fuel.pack(pady=5)

tk.Label(frame, text="Transport Usage (km):", font=("Arial", 24, "bold")).pack(pady=5)
entry_transport = tk.Entry(frame, font=("Arial", 24), justify="center", width=15)
entry_transport.pack(pady=5)

tk.Button(frame, text="Calculate", command=calculate_carbon_footprint, 
          font=("Arial", 20, "bold"), bg="lightblue", padx=20).pack(pady=10)
tk.Button(frame, text="Reset", command=reset_fields, 
          font=("Arial", 20, "bold"), bg="lightyellow", padx=20).pack(pady=10)
tk.Button(frame, text="Exit", command=confirm_exit, 
          font=("Arial", 20, "bold"), bg="lightcoral", padx=20).pack(pady=10)

label_result = tk.Label(frame, text="", font=("Arial", 28, "bold"), 
                       fg="darkgreen", bg="lightgray", 
                       wraplength=800, pady=15, padx=20)
label_result.pack(pady=20, fill='x')

print("Application started! Enter values and click Calculate.")
root.mainloop()
```
